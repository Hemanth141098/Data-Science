```python
# Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
file_path = 'E:/Data Science Project/Unified Mentor Project/Intership Project/Project 8_ Employee Attrition Analysis/Attrition data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Understanding the structure of the dataset
data.info()

# Statistical summary of numerical columns
data.describe()
```

```python
# Handling missing values
data.dropna(inplace=True)

# Removing irrelevant columns
irrelevant_columns = ['EmployeeID', 'EmployeeCount', 'Over18', 'StandardHours']
data.drop(columns=irrelevant_columns, inplace=True)

# Confirming the changes
data.info()
```

```python
print("\nMissing values after cleaning:")
print(data.isnull().sum())
```

```python
# Visualizing distributions of numerical columns
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')

plt.subplot(2, 2, 2)
sns.boxplot(x='Department', y='MonthlyIncome', data=data)
plt.title('Monthly Income Distribution by Department')

plt.subplot(2, 2, 3)
sns.countplot(x='Attrition', data=data, palette='Set2')
plt.title('Attrition Count')

plt.subplot(2, 2, 4)
sns.scatterplot(x='TotalWorkingYears', y='YearsAtCompany', data=data, hue='Attrition')
plt.title('Years at Company vs. Total Working Years')

plt.tight_layout()
plt.show()
```

```python
# Histograms and distributions of numerical variables
plt.figure(figsize=(14, 12))

plt.subplot(3, 2, 1)
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')

plt.subplot(3, 2, 2)
sns.histplot(data['MonthlyIncome'], bins=20, kde=True)
plt.title('Distribution of Monthly Income')

plt.subplot(3, 2, 3)
sns.histplot(data['TotalWorkingYears'], bins=20, kde=True)
plt.title('Distribution of Total Working Years')

plt.subplot(3, 2, 4)
sns.histplot(data['YearsAtCompany'], bins=20, kde=True)
plt.title('Distribution of Years at Company')

plt.subplot(3, 2, 5)
sns.histplot(data['JobSatisfaction'], bins=5, kde=True)
plt.title('Distribution of Job Satisfaction')

plt.subplot(3, 2, 6)
sns.histplot(data['WorkLifeBalance'], bins=5, kde=True)
plt.title('Distribution of Work-Life Balance')

plt.tight_layout()
plt.show()
```

```python
# Pairplot for numerical variables
sns.pairplot(data[['Age', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany', 'JobSatisfaction', 'WorkLifeBalance']])
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Selecting only numerical columns for correlation
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data[numerical_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# Count plots for categorical variables
plt.figure(figsize=(14, 10))

plt.subplot(3, 2, 1)
sns.countplot(x='Attrition', data=data, palette='Set1')
plt.title('Attrition Count')

plt.subplot(3, 2, 2)
sns.countplot(x='BusinessTravel', data=data, palette='Set2')
plt.title('Business Travel Frequency')

plt.subplot(3, 2, 3)
sns.countplot(x='Department', data=data, palette='Set3')
plt.title('Department Distribution')

plt.subplot(3, 2, 4)
sns.countplot(x='EducationField', data=data, palette='pastel')
plt.title('Education Field Distribution')

plt.subplot(3, 2, 5)
sns.countplot(x='Gender', data=data, palette='deep')
plt.title('Gender Distribution')

plt.subplot(3, 2, 6)
sns.countplot(x='MaritalStatus', data=data, palette='muted')
plt.title('Marital Status Distribution')

plt.tight_layout()
plt.show()
```
